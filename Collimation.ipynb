```python
#import

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

#parameter
lam = 852e-9 #wave length[m]
```

```python
zs = np.array([5,6,8,10,12,14,16,30,32,34,36,38,40,42,44])*25*1e-3 #[m]
ws = np.array([1590,1580,1620,1670,1680,1735,1770,2070,2100,2130,2170,2180,2270,2330,2360])/2*1e-6 #[m]

#任意の位置:zにおけるビーム径を
def waist(z, z0, w0):
    z_R = np.pi*1*w0**2/lam #Rayleigh length
    return w0*np.sqrt(1 + ((z-z0)/z_R)**2)

z = np.linspace(-3000e-3, 3000e-3, 6001)

popt, pcov = curve_fit(waist, zs, ws)
waist_position, Omega0 = popt
print("rayleigh length:",np.pi*1*popt[1]**2/lam,"[m]")
print("waist position:",waist_position,"[m]")
print("beam waist:",Omega0,"[m]")



plt.figure(figsize=(8,4))
plt.plot(z, waist(z, *popt))
plt.plot(zs, ws, '.')
plt.xlabel("length[m]")
plt.ylabel("radius[m]")
plt.show()
```
